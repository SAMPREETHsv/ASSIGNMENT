# PRIME OR NOT



```python
num=int(input("Enter a number      "))
if(num>1):
  for i in range(2,int(num/2)+1):
    if(num%i==0):
      print("The number ",num," is not a prime number.")
      break
    else:
      print(num, " is a prime number.")
      break
else:
  print("Number is not prime number. ")
```

# FIBONACCI SERIES

```python
num=int(input("enter the fibonacci length :    "))
firsterm=0
secondterm=1
print("The fibonacci series with ", num, "  terms is:  ")
print(firsterm,secondterm,end=" ")
for i in range(2,num):
  curterm=firsterm+secondterm
  print(curterm, end=" ")
  firsterm=secondterm
  secondterm=curterm
```

# FACTORIAL OF A NUMBER

```python
def fact(num):
  if num==0:
    return 1
  else:
    return num*fact(num-1)

a=int(input("Enter the value for factorial:  "))
print("Factorial of ",a," is:  ",fact(a))
```

# REVERSE A NUMBER USING STRING

```python
num=int(input("Enter the number"))
reversed_num=0
while num!=0:
  digit=num%10
  reversed_num=reversed_num*10+digit
  num//=10
  print("Reversed number=   "+ str(reversed_num))
```
